一般来说处理空值有以下几种方法：
1.少量空值可以根据前后信息进行填补，但一般来说能被较好补上的信息说明没有被补上的必要，不能被较好的补上的信息补出来的结果可能有人不是很好，一般来说介于两者之间；
2.用平均值、中值、分位数、众数、随机值等替代。效果一般，因为等于人为增加了噪声；
3.最精确的做法：把变量映射到高维空间，把男，女，缺失映射为，是否男，是否女，是否缺失。样本量大的时候效果会比较好，否则会出现稀疏；
4.缺失量实在过大则直接删除。

```python
df['Second_Assistant_ID'] = df['Second Assistant ID'].apply(lambda x:x )
```

```python
df['Second_Assistant_ID'] = df['Second_Assistant_ID'].apply(lambda x: pd.to_numeric(x, errors='coerce')) 
```

```python
from sklearn.ensemble import RandomForestRegressor
#用random forest来拟合填充
def set_missing_SA(df):
    #df = df.reset_index()
    #把数值型的特征取出放到regressor里面
    process_df = df[['Second_Assistant_ID','Surgeon ID','Anesthesiologist ID', 'Circulator ID']]
    #把预测值分成已知和未知两部分
    know = process_df[process_df.Second_Assistant_ID.notnull()].as_matrix()
    unknow = process_df[process_df.Second_Assistant_ID.isnull()].as_matrix()
    # X为特征属性值
    X = know[:, 1:]
    # y为结果标签值
    y = know[:, 0]
    rfr = RandomForestRegressor(random_state = 0,n_estimators = 2000,n_jobs=-1)
    rfr.fit(X,y)
    #用模型来预测未知量
    predicted = rfr.predict(unknow[:,1::])
    #用得到的预测结果填补原缺失数据
    df.loc[(df.Second_Assistant_ID.isnull(),'Second_Assistant_ID')] = predicted
    return df, rfr
df,rfr = set_missing_SA(df)
```
